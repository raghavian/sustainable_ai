### Figure 1-1: Historical trend of global temperature anomaly and sea levels. (Source: [Two Degrees Institute](https://2degreesinstitute.org))

```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib

params = {'font.size': 16,
          'font.weight': 'normal',
          'axes.labelsize':16,
          'axes.titlesize':16,
          'axes.labelweight':'normal',
          'axes.titleweight':'normal',
          'legend.fontsize': 16,
          "text.usetex": True,
         }
matplotlib.rcParams.update(params)

# Load the CSV files
temp_df = pd.read_csv('../data/ch01_temp_2025-07-13.csv')
sea_df = pd.read_csv('../data/ch01_sea_2025-07-13.csv')

# Extract year from the date strings
temp_df['Year'] = temp_df['Temperature Date'].str[:4].astype(int)
sea_df['Year'] = sea_df['Sealevels Date'].str[:4].astype(int)

# Create a shared x-axis plot with twin y-axes
fig, ax1 = plt.subplots(figsize=(16, 6))

# Plot temperature values with solid line and circle markers
line1, = ax1.plot(temp_df['Year'], temp_df['Temperature Value'], linestyle='-', color='tab:red', label='Temperature Anomaly')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature Anomaly (C)', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create a second y-axis for sea level values
ax2 = ax1.twinx()
line2, = ax2.plot(sea_df['Year'], sea_df['Sealevels Value'], linestyle='--', color='tab:blue', label='Sea Level')
ax2.set_ylabel('Sea Level (cm)', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add a combined legend
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

plt.grid(True)
fig.tight_layout()
plt.show()
```

### Figure 1-7. Research trend on AI and sustainability over the last two decades. (Source: Data from [Dimensions.ai](https://app.dimensions.ai) using the search keywords AI and sustainability.)

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
params = {'font.size': 16,
          'font.weight': 'normal',
          'axes.labelsize':16,
          'axes.titlesize':16,
          'axes.labelweight':'normal',
          'axes.titleweight':'normal',
          'legend.fontsize': 16,
          "text.usetex": True,
         }
matplotlib.rcParams.update(params)
df = pd.read_csv('../data/ch01_publications.csv')

# Plotting the number of publications over time
plt.figure(figsize=(16, 6))
plt.plot(df['Category'], df['Publications (total)'], marker='o', linewidth=4)
plt.fill_between(df['Category'], df['Publications (total)'], color='tab:blue', alpha=0.3)
plt.ylabel('Number of Publications')
plt.grid(True)
# Set x-ticks to appear every 3 years
years = df['Category']
plt.xticks(ticks=years[::2], rotation=45)
plt.tight_layout()
plt.show()
```
